# Fetching Biometric Data Through Wearables

This notebook explores [Sahha's](https://sahha.ai/) API for fetching wearable device data. For more information about the API, please refer to the [documentation](https://docs.sahha.ai/). This can be considered as a proof of concept for further development and/or investigation.

```python
# importing dependencies


import os
import json
import requests
from dotenv import load_dotenv

# sahha_api module implements some api endpoints with the requests library
# please, refer to `sahha_api.py` to check the implementation
from sahha_api import get_auth_token, get_steps, get_demographic, get_device_information, get_sleep, get_health
```

```python
# getting credentials from .env file

load_dotenv()

CLIENT_ID = os.environ["CLIENT_ID"]
CLIENT_SECRET = os.environ["CLIENT_SECRET"]
PATIENT_ID = os.environ["PATIENT_ID"]
```

## Examples

Here are some API usage examples:

### Authenticating

```python
# getting authorization token from the api 
account_token = get_auth_token(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
```

### Getting Device Information

```python
device_info = get_device_information(token=account_token, user_id=PATIENT_ID)

print(json.dumps(device_info, indent=4))
```

### Demographic

The api can also produce some demographic data provided by the user through the mobile app

```python
demographic_data = get_demographic(token=account_token, user_id=PATIENT_ID)

print(json.dumps(demographic_data, indent=4))
```

### Steps

The api can fetch the number of steps
This could be useful for measuring patient/user activity

```python
steps = get_steps(token=account_token, user_id=PATIENT_ID)

print(json.dumps(steps, indent=4))
```

### Fetching Sleep Data

Another interesting dimension that may provide insightful correlations is sleep data

```python
sleep_data = get_sleep(token=account_token, user_id=PATIENT_ID)

print(json.dumps(sleep_data, indent=4))
```

```python
health_data = get_health(token=account_token, user_id=PATIENT_ID, datetime="2025-04-20")

print(json.dumps(health_data, indent=4))
```
